```python
%%sql
CREATE VIEW facts_exclude_world AS
SELECT *
  FROM facts
 WHERE name <> 'World';
```

```python
%%sql
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
  FROM facts_exclude_world
```

```python
# Average Population and area based on the new factbook
```

```python
%%sql
SELECT AVG(population) as average_pop, 
       AVG(area) as average_area
  FROM facts_exclude_world;
```

```python
%%sql
SELECT *
  FROM facts_exclude_world
 WHERE population > ( SELECT AVG(population) 
                     FROM facts_exclude_world)
 AND
    area > ( SELECT AVG(area)
             FROM facts_exclude_world)
```

```python
# Country with highest growth rate
```

```python
%%sql
SELECT name, population_growth, birth_rate, death_rate
  FROM facts_exclude_world
    WHERE population_growth = (SELECT 
                                   MAX(population_growth)
                               FROM
                                   facts_exclude_world)
```

```python
# highest ratios of water to land and countries that have more water than land
```

```python
%%sql
SELECT name, population_growth, birth_rate, migration_rate
  FROM facts_exclude_world
    WHERE migration_rate = (SELECT 
                                   MAX(migration_rate)
                               FROM
                                   facts_exclude_world)
```

```python
# highest death rate than birth rate
```

```python
%%sql
SELECT name, population_growth, birth_rate, death_rate
  FROM facts_exclude_world
    WHERE birth_rate < death_rate;
```
